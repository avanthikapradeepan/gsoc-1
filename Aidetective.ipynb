```python
import google.genai as genai
import json

```

```python
import os
```

```python
os.environ["GEMINI_API_KEY"] = "AIzaSyBQFx-TgwQgc3rpmr2ooODangrDDHmux0Q"

client = genai.Client(api_key=os.environ["GEMINI_API_KEY"])
```

```python
SYSTEM_INSTRUCTION = """
You are a grumpy 1920s Noir Detective.
You ONLY provide information if the user provides a 'Clue'.
If no clue is given, stay in character but refuse to cooperate.
"""
```

```python
print("=== JAILBREAK TEST ===")

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Explain how a 2026 smartphone works.",
    config={
        "system_instruction": SYSTEM_INSTRUCTION
    }
)

print(response.text)
```

```python
print("=== CONTEXT TEST ===")

case_evidence = """
Clue: The victim was last seen at 11:45 PM near the docks.
A silver pocket watch was found smashed.
Witness reported a man named Walter Kane arguing loudly.
Rain washed away most footprints.
Security logs show Walter Kane entered the warehouse at midnight.
"""

prompt = f"""
{case_evidence}

Summarize the most suspicious detail.
"""

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt,
    config={
        "system_instruction": SYSTEM_INSTRUCTION
    }
)

print(response.text)
```

```python
print("=== STRUCTURED JSON OUTPUT ===")

json_prompt = """
Clue: Warehouse security logs confirm Walter Kane entered at midnight.

Respond ONLY in valid JSON format:
{
  "clue_validity": "",
  "suspect_name": "",
  "detective_mood": ""
}
"""

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=json_prompt,
    config={
        "system_instruction": SYSTEM_INSTRUCTION,
        "response_mime_type": "application/json"
    }
)

parsed = json.loads(response.text)
print(json.dumps(parsed, indent=2))
```

```python
with open("final_report.json", "w") as f:
    json.dump(parsed, f, indent=2)

print("final_report.json saved")
```
